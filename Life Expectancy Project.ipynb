```python
df.dropna(subset=['Life expectancy '],axis=0, inplace=True)
df['Life expectancy '].isnull().sum()
```

# Impute Alcohol column

```python
import numpy as np

# Get list of countries with nulls in 'Alcohol'
countriesAl_with_nulls = df[df['Alcohol'].isnull()]['Country'].unique()

# Impute missing values with country-specific mean
for country in countriesAl_with_nulls:
    country_mean = df.loc[df['Country'] == country, 'Alcohol'].mean()
    df.loc[(df['Country'] == country) & (df['Alcohol'].isnull()), 'Alcohol'] = country_mean
```

```python
df['Alcohol'].isnull().sum()
```

```python
m=df['Alcohol'].mean()
df['Alcohol'].fillna(m,inplace=True)
df['Alcohol'].isnull().sum()
```

# Trends of 'Life expectancy ' of each country over years

```python
df=pd.read_csv("Life Expectancy Data.csv")
fig, axes = plt.subplots(2, 5, figsize=(24, 10), sharey=True)

# Flatten axes for easy iteration
axes = axes.flatten()

# Loop through the first 10 countries and plot each on a subplot
for idx, country in enumerate(countries_List[:10]):
    ax = axes[idx]
    df_country = df[df['Country'] == country][['Year', 'Life expectancy ']].sort_values('Year')
    
    # Plot on the respective axis
    ax.plot(df_country['Year'], df_country['Life expectancy '], marker='o', linewidth=2.5)
    ax.set_title(f"{country}", fontsize=12)
    ax.set_xlabel("Year")
    if idx % 5 == 0:
        ax.set_ylabel("Life Expectancy")
    ax.grid(True)

# Adjust layout
plt.suptitle("Life Expectancy Trends for Top 10 Countries", fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import folium
#Create a plain worl map
# Create the map object
world_map = folium.Map(zoom_start=2, tiles='cartodbpositron')  # 'Mapbox Bright' is no longer supported

# GeoJSON file path
world_geo = r'countries.geo.json'

# Create the Choropleth map
folium.Choropleth(
    geo_data=world_geo,
    name='choropleth',
    data=df_map,
    columns=['Country', 'Life expectancy '],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Life Expectancy'
).add_to(world_map)

# Add layer control
folium.LayerControl().add_to(world_map)

# Display the map
world_map
```
